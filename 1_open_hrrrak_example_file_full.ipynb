Merging this list into a simple dataset.

Keeping in mind there is data variable overlap for height, so I'm using override here for a quick look but this is not everything.

```python
# now merge all the datasets from ds_list 
ds_all = xr.merge(ds_list, compat='override')

ds_all
```

Calculate wind speed, 

```python
# calculate wind speed from u and v
ds_all['wspd'] = np.sqrt(ds_all['u10']**2 + ds_all['v10']**2)
ds_all['wspd'].plot()
```
